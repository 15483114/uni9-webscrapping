```python

import requests
from bs4 import BeautifulSoup
from bs4 import Comment
import pandas as pd

url = 'https://www.basketball-database.com/csgc/league/nbb/3526#tab4'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
```
